```python
import requests
import yaml
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10,8
```

```python
with open("conf.yml", "r+") as f:
    conf = yaml.load(f)
```

```python
API_KEY = conf['API_KEY']
```

```python
STR_URL_FMT = 'https://www.materialsproject.org/rest/v2/materials/{element}/vasp/structure?API_KEY={API_KEY}'
```

```python
res = requests.get(STR_URL_FMT.format(element='Cu', API_KEY=API_KEY))
```

```python
res.json()['response']
```

```python
res.json()['response'][0]
```

```python
res.json()['response'][0]['structure']['lattice']
```

```python
res.json()['response'][0]['structure']['lattice']['matrix']
```

```python
mat = res.json()['response'][0]['structure']['lattice']['matrix']
```

```python
x = []
y = []
z = []

for i in range(-3, 4):
    for j in range(-3, 4):
        for k in range(-3, 4):
            x.append(i*np.array(mat[0]))
            y.append(j*np.array(mat[1]))
            z.append(k*np.array(mat[2]))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x, y, z)
plt.show()
```

```python
x = []
y = []
z = []

for i in range(2):
    for j in range(2):
        for k in range(2):
            x.append(i*np.array(mat[0]))
            y.append(j*np.array(mat[1]))
            z.append(k*np.array(mat[2]))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x, y, z)
plt.show()
```
